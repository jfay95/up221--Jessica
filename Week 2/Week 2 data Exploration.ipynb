Next, I will examine the count of SNAP recipients in each SNAP Tract with the .values_counts() funciton. Then I save the output as a variable in order to turn it into a data frame that I can use to plot.

```python
# I am printing the count of SNAP Recipients in each SNAPTract
usda['TractSNAP'].value_counts()

#save as a variable
snap_count=usda['TractSNAP'].value_counts()
```

```python
#I am converting the series to a panda data frame
snap_count=snap_count.reset_index()
snap_count
```

Above is the print out of the new data frame of the count of recipients per SNAP tract. The highest count is 96 for SNAP tract 0, which could maybe indicate missingness. 
Next I will subset the data to make it simpler for plotting. I am going to plot count of recipients per SNAP tract but since some tract numbers are so high it will make the bar chart hard to read. So for ease of readability, I am going to plot 3 SNAP tract numbers only. I will filter out the dataframe for 3 tracts and plot the count of recipients.

```python
# I am subsetting the data to make it simpler for practicing graphing, if I don't subset it there are 338 x values. Here I am filtering for only SNAP tracts of 42, 53 and 10.

graph = snap_count[snap_count["TractSNAP"].isin(["53","42","10"])]
graph

```

Above is the filtered data frame, where tract 42 had 10 recipients, tract 53 had 9 recipients and tract 10 had 8 recipients. 
Next, I will plot these counts in a barchart.

```python
#plot the subsetted data
graph.plot.bar(x='TractSNAP', y ='count', title= 'Number of SNAP Recipients per SNAP Tract')
```
