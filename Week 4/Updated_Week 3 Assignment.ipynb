```python
#maps, import geopandas
import geopandas as gpd
```

## The code will be used to join the census tract data with the age data using FIPS as common var

```python
# read in a geojson file downloaded from the LA Times
tracts=gpd.read_file('Data/Census_Tracts_2020.geojson')
tracts.head()
```

```python
tracts.plot(figsize=(12,10))
```

```python
# subset to get FIPS and geometry
tracts = tracts[['CT20','geometry']]
tracts.head()
```

```python
# create a FIPS column to merge
tracts['FIPS'] ='06' + '037' + tracts['CT20']
```

```python
tracts.head()
```

```python
#join tracts and census data
tracts_age=tracts.merge(df2,on="FIPS")
tracts_age.head()
```

## The next maps will show the spatial distribution of population aged 35-39, under 5 and over 85

```python
tracts_age.plot(figsize=(12,10),
                 column='35 to 39',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
tracts_age.plot(figsize=(12,10),
                 column='Under 5',
                 legend=True, 
                 scheme='equal_interval')
```

```python
tracts_age.plot(figsize=(12,10),
                 column='85 and over',
                 legend=True, 
                 scheme='equal_interval')
```

## Now I will use folium to map spatial distribution of those under 5

```python
import folium 
m = folium.Map(location=[34.2,-118.2], 
               zoom_start = 9,
               tiles='CartoDB positron', 
               attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
                  geo_data=tracts_age, # geo data
                  data=tracts_age, # data          
                  key_on='feature.properties.FIPS', # key, or merge column
                  columns=['FIPS', 'Under 5'], # [key, value]
                  fill_color='BuPu',
                  line_weight=0.1, 
                  fill_opacity=0.8,
                  line_opacity=0.2, # line opacity (of the border)
                  legend_name='Population Under 5 (2020)').add_to(m)    # name on the legend color bar

```

```python
m.save('LA_underfive.html')
```
